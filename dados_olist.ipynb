```python
import urllib.request

url_orders = "https://raw.githubusercontent.com/fiesc-junior-prado/mine_projeto_bloco_1/refs/heads/main/olist_orders_dataset.csv"
with urllib.request.urlopen(url_orders) as response:
    orders_data = response.read().decode('utf-8')
    

url_products = "https://raw.githubusercontent.com/fiesc-junior-prado/mine_projeto_bloco_1/refs/heads/main/olist_products_dataset.csv"
with urllib.request.urlopen(url_products) as response:
    products_data = response.read().decode('utf-8')
    
```

```python
# Validação e Tratamento de Dados Ausentes:

import csv
import io
from tabulate import tabulate

# O arquivo já foi baixado antes, então vamos usar o conteúdo dele
# products_data é o conteúdo do arquivo CSV que já temos

# Passo 1: Ler o conteúdo do CSV
leitor = csv.reader(io.StringIO(products_data))
todas_linhas = list(leitor)

# Passo 2: Pegar o cabeçalho (primeira linha)
cabeçalho = todas_linhas[0]

# Passo 3: Descobrir em qual coluna está cada coisa que queremos
# Vamos procurar a coluna de categoria
indice_categoria = -1
for posicao, nome_coluna in enumerate(cabeçalho):
    if nome_coluna == 'product_category_name':
        indice_categoria = posicao
        break

# Vamos procurar as colunas de peso e medidas
indice_peso = -1
indice_comprimento = -1
indice_altura = -1
indice_largura = -1

for posicao, nome_coluna in enumerate(cabeçalho):
    if nome_coluna == 'product_weight_g':
        indice_peso = posicao
    elif nome_coluna == 'product_length_cm':
        indice_comprimento = posicao
    elif nome_coluna == 'product_height_cm':
        indice_altura = posicao
    elif nome_coluna == 'product_width_cm':
        indice_largura = posicao

# Passo 4: Calcular a média de cada medida (só dos valores que existem)
soma_peso = 0
quantidade_peso = 0

soma_comprimento = 0
quantidade_comprimento = 0

soma_altura = 0
quantidade_altura = 0

soma_largura = 0
quantidade_largura = 0

# Percorrer todas as linhas (começando da 1, pois a 0 é o cabeçalho)
for numero_linha, linha in enumerate(todas_linhas):
    if numero_linha == 0:
        continue  # Pula o cabeçalho
    
    # Verificar peso
    if indice_peso >= 0 and len(linha) > indice_peso:
        valor = linha[indice_peso].strip()
        if valor != '' and valor.lower() != 'null':
            try:
                soma_peso = soma_peso + float(valor)
                quantidade_peso = quantidade_peso + 1
            except:
                pass  # Se der erro, ignora
    
    # Verificar comprimento
    if indice_comprimento >= 0 and len(linha) > indice_comprimento:
        valor = linha[indice_comprimento].strip()
        if valor != '' and valor.lower() != 'null':
            try:
                soma_comprimento = soma_comprimento + float(valor)
                quantidade_comprimento = quantidade_comprimento + 1
            except:
                pass
    
    # Verificar altura
    if indice_altura >= 0 and len(linha) > indice_altura:
        valor = linha[indice_altura].strip()
        if valor != '' and valor.lower() != 'null':
            try:
                soma_altura = soma_altura + float(valor)
                quantidade_altura = quantidade_altura + 1
            except:
                pass
    
    # Verificar largura
    if indice_largura >= 0 and len(linha) > indice_largura:
        valor = linha[indice_largura].strip()
        if valor != '' and valor.lower() != 'null':
            try:
                soma_largura = soma_largura + float(valor)
                quantidade_largura = quantidade_largura + 1
            except:
                pass

# Calcular as médias
if quantidade_peso > 0:
    media_peso = soma_peso / quantidade_peso
else:
    media_peso = 0

if quantidade_comprimento > 0:
    media_comprimento = soma_comprimento / quantidade_comprimento
else:
    media_comprimento = 0

if quantidade_altura > 0:
    media_altura = soma_altura / quantidade_altura
else:
    media_altura = 0

if quantidade_largura > 0:
    media_largura = soma_largura / quantidade_largura
else:
    media_largura = 0

# Passo 5: Criar novo arquivo com os dados corrigidos
arquivo_saida = open("produtos_processados.csv", "w", encoding="utf-8", newline="")
escritor = csv.writer(arquivo_saida)

for numero_linha, linha in enumerate(todas_linhas):
    if numero_linha == 0:
        # Escreve o cabeçalho igual
        escritor.writerow(linha)
        continue
    
    # Corrigir categoria se estiver vazia
    if indice_categoria >= 0:
        # Garantir que a linha tem essa coluna
        while len(linha) <= indice_categoria:
            linha.append("")
        
        valor_categoria = linha[indice_categoria].strip()
        if valor_categoria == '' or valor_categoria.lower() == 'null':
            linha[indice_categoria] = "Sem categoria"
    
    # Corrigir peso se estiver vazio
    if indice_peso >= 0:
        while len(linha) <= indice_peso:
            linha.append("")
        valor = linha[indice_peso].strip()
        if valor == '' or valor.lower() == 'null':
            linha[indice_peso] = str(round(media_peso, 2))
    
    # Corrigir comprimento se estiver vazio
    if indice_comprimento >= 0:
        while len(linha) <= indice_comprimento:
            linha.append("")
        valor = linha[indice_comprimento].strip()
        if valor == '' or valor.lower() == 'null':
            linha[indice_comprimento] = str(round(media_comprimento, 2))
    
    # Corrigir altura se estiver vazio
    if indice_altura >= 0:
        while len(linha) <= indice_altura:
            linha.append("")
        valor = linha[indice_altura].strip()
        if valor == '' or valor.lower() == 'null':
            linha[indice_altura] = str(round(media_altura, 2))
    
    # Corrigir largura se estiver vazio
    if indice_largura >= 0:
        while len(linha) <= indice_largura:
            linha.append("")
        valor = linha[indice_largura].strip()
        if valor == '' or valor.lower() == 'null':
            linha[indice_largura] = str(round(media_largura, 2))
    
    # Escrever a linha corrigida
    escritor.writerow(linha)

# Fechar o arquivo
arquivo_saida.close()

# Mostrar mensagem de sucesso
print("Arquivo processado salvo como 'produtos_processados.csv'")

# Visualizar as primeiras linhas como tabela
print(tabulate(todas_linhas[:11], headers="firstrow", tablefmt="grid"))
```

```python
# Padronização de Strings e Regex:

import csv
import io
import re
from tabulate import tabulate

# Ler o conteúdo do CSV
leitor = csv.reader(io.StringIO(products_data))
todas_linhas = list(leitor)

# Pegar o cabeçalho e encontrar o índice da coluna de categoria
cabeçalho = todas_linhas[0]
indice_categoria = cabeçalho.index('product_category_name')

# Processar cada linha (exceto o cabeçalho)
for i in range(1, len(todas_linhas)):
    linha = todas_linhas[i]
    
    # Pegar o valor da categoria
    categoria = linha[indice_categoria]
    
    # 1. Remover espaços em branco no início e no fim
    categoria = categoria.strip()
    
    # 2. Converter tudo para minúsculas
    categoria = categoria.lower()
    
    # 3. Remover caracteres especiais e pontuações com Regex
    categoria = re.sub(r'[^a-z0-9\s]', '', categoria)
    
    # 4. Remover espaços extras que sobraram
    categoria = categoria.strip()
    
    # 5. Se ficou vazio, colocar "sem categoria"
    if categoria == "":
        categoria = "sem categoria"
    
    # Atualizar a linha
    linha[indice_categoria] = categoria

# Salvar o arquivo processado
with open("produtos_processados.csv", "w", encoding="utf-8", newline="") as f_out:
    escritor = csv.writer(f_out)
    for linha in todas_linhas:
        escritor.writerow(linha)

print("Arquivo processado salvo como 'produtos_processados.csv'")

# Visualizar as primeiras linhas como tabela
print(tabulate(todas_linhas[:11], headers="firstrow", tablefmt="grid"))
```
