```python
# Lógica de Regra de Negócio (Filtros e Validação)

import csv
import urllib.request

url_orders = "https://raw.githubusercontent.com/fiesc-junior-prado/mine_projeto_bloco_1/refs/heads/main/olist_orders_dataset.csv"
caminho_arquivo = "orders.csv"
urllib.request.urlretrieve(url_orders, caminho_arquivo)

with open(caminho_arquivo, "r", encoding="utf-8") as f:
    leitor = csv.reader(f)
    linhas = list(leitor)

# Analisar os pedidos para verificar a hipótese sobre a relação entre status e data de entrega
cabeçalho = linhas[0]
idx_status = cabeçalho.index("order_status")
idx_entrega = cabeçalho.index("order_delivered_customer_date")

total_pedidos = 0
pedidos_sem_entrega = 0
cancelados_sem_entrega = 0
nao_cancelados_sem_entrega = 0

# Percorrer os pedidos e contar os casos relevantes para a análise
for i in range(1, len(linhas)):
    linha = linhas[i]
    total_pedidos += 1
    status = linha[idx_status].strip()
    data_entrega = linha[idx_entrega].strip()
    if data_entrega == "" or data_entrega.lower() == "null":
        pedidos_sem_entrega += 1
        if status == "canceled":
            cancelados_sem_entrega += 1
        else:
            nao_cancelados_sem_entrega += 1

# Exibir os resultados da análise
print(f"Total de pedidos analisados: {total_pedidos}")
print(f"Pedidos sem data de entrega: {pedidos_sem_entrega}")
print(f"  - Cancelados: {cancelados_sem_entrega}")
print(f"  - NÃO Cancelados: {nao_cancelados_sem_entrega}")
if nao_cancelados_sem_entrega == 0 and pedidos_sem_entrega > 0:
    print("\n Hipótese CONFIRMADA: Todos os pedidos sem data de entrega estão cancelados.")
elif pedidos_sem_entrega == 0:
    print("\n Não existem pedidos sem data de entrega para analisar.")
else:
    print(f"\n Hipótese REJEITADA: Existem {nao_cancelados_sem_entrega} pedidos sem data de entrega que NÃO foram cancelados.")
```
