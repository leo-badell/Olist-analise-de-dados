```python
# Relatório de Status Manual

import csv
import os

# Lê o CSV e conta as linhas, categorias corrigidas e nulos corrigidos
def contar_sumario_produtos(caminho_arquivo: str) -> tuple[int, int, int]:
    """Lê o CSV e retorna (total_linhas, categorias_corrigidas, nulos_corrigidos)."""
    COLUNAS_DIMENSAO = [
        'product_weight_g',
        'product_length_cm',
        'product_height_cm',
        'product_width_cm',
    ]

    contagem_total = 0
    nulos_corrigidos = 0
    categorias_corrigidas = 0
# Abrir o arquivo CSV e processar linha por linha
    with open(caminho_arquivo, "r", encoding="utf-8") as f:
        leitor = csv.DictReader(f)
        for linha in leitor:
            contagem_total += 1

            if linha.get('product_category_name', '').strip() == "sem categoria":
                categorias_corrigidas += 1

            for col in COLUNAS_DIMENSAO:
                valor = linha.get(col, '').strip().lower()
                if valor in ('', 'null', 'none'):
                    nulos_corrigidos += 1

    return contagem_total, categorias_corrigidas, nulos_corrigidos

# Exibe o relatório de status formatado no console
def mostrar_sumario(total: int, categorias: int, nulos: int) -> None:
    """Exibe o relatório de status formatado no console."""
    separador = "-" * 45
    print(separador)
    print("  RELATÓRIO DE STATUS — produtos_processados")
    print(separador)
    print(f"  Total de linhas processadas       : {total}")
    print(f"  Categorias corrigidas (sem categ.): {categorias}")
    print(f"  Campos de dimensões nulos         : {nulos}")
    print(separador)

CAMINHO_ARQUIVO = os.path.join(os.path.dirname(os.path.abspath("__file__")), "produtos_processados.csv")

total, categorias, nulos = contar_sumario_produtos(CAMINHO_ARQUIVO)
mostrar_sumario(total, categorias, nulos)
```
